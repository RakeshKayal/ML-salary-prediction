```python
import pandas as pd


from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import joblib

# Load dataset
d = pd.read_csv('regression_dataset.csv')
print(d.head(3))

```

```python
d.isnull().sum() #so there is no null value
```

```python
d.describe() #mean and max of Salary is nearly  to each Other
```

```python
# just see the graph to determine which Algo use in this SuperVised Learning

import matplotlib.pyplot as plt


plt.scatter(d["Experience"], d["Salary"], c=d["Age"], cmap="viridis", label="Data")
plt.colorbar(label="Age")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience (color = Age)")
plt.show()
# see it give as age an experience increase salary increage.  so it need multiple Regression.. 
```

```python
 x = d.iloc[:, :-1]
 y = d["Salary"]
 print(x.head(3)) # select the Feature  
```

```python
print (y.head(3))
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

 # Train model
 lr = LinearRegression()
 lr.fit(x_train, y_train)
```

```python
score = lr.score(x_test, y_test) * 100
print(f"Model Accuracy: {score:.2f}%")
```

```python
print("Intercept:", lr.intercept_)
print("Coefficients:", lr.coef_)
```
